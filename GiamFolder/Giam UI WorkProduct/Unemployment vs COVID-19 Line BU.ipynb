```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
from scipy.stats import linregress
from pprint import pprint
import json
from datetime import datetime 

# Pull source UI file
ui_cv_line_df = pd.read_csv("../Resources/ui_cv_scat_clean.csv")
ui_cv_line_df.head(1)
```

```python
cal = ui_cv_line_df[ui_cv_line_df["State"] == 'California']
cal.head(1)
```

```python
t = cal["Weeknum"]
data1 = cal["Insured Unemployment Rate"]
data2 = cal["COVID-19"] 

fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('Week Number')
ax1.set_ylabel('% Unemployment', color=color)
ax1.plot(t, data1, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel("COVID-19", color=color)  # we already handled the x-label with ax1
ax2.plot(t, data2, color=color)
ax2.tick_params(axis='y', labelcolor=color)
plt.title("California Unemployment vs COVID-19 Positive Increases")

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.savefig('../output_data/UI_CV_Line/Cal_UI_CV.png')
plt.show()
```
