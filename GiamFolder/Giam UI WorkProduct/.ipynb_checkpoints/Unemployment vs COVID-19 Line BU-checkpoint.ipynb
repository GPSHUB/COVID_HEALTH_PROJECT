```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
from scipy.stats import linregress
from pprint import pprint
import json
from datetime import datetime 

# Pull source UI file
ui_cv_line_df = pd.read_csv("../Resources/ui_cv_scat_clean.csv")
ui_cv_line_df.head(1)
```

```python
cal = ui_cv_line_df[ui_cv_line_df["State"] == 'California']
cal.head(1)
```

```python
cal = ui_cv_line_df[ui_cv_line_df["State"] == 'California']

plt.title("California Unemployment vs COVID-19 Positive Increases")
plt.xlabel("Week Number")
plt.ylabel("% Unemployment / Anxiety Depression")

x_values = cal["Weeknum"]
y_values = cal[["Insured Unemployment Rate", "COVID-19"]]

plt.plot(x_values, y_values)

# plt.ylim(0,10000)

plt.savefig('../output_data/UI_CV_Line/Cal_UI_CV_Line.png')

plt.show()
```
