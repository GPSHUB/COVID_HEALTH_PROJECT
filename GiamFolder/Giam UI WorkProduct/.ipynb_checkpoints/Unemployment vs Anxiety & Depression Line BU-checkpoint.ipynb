```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
from scipy.stats import linregress
from pprint import pprint
import json
from datetime import datetime 

# Pull source UI file
ui_ad_line_df = pd.read_csv("../Resources/ui_ad_scat_clean.csv")
ui_ad_line_df
```

```python
cal = ui_ad_line_df[ui_ad_line_df["State"] == 'California']
cal
```

```python
cal = ui_ad_line_df[ui_ad_line_df["State"] == 'California']

plt.title("California Unemployment vs Anxiety & Depression")
plt.xlabel("Week Number")
plt.ylabel("% Unemployment / Anxiety Depression")

x_values = cal["Weeknum"]
y_values = cal[["Insured Unemployment Rate", "Anxiety"]]

plt.plot(x_values, y_values)
plt.ylim(0,47)

plt.savefig('../output_data/UI_AD_Line/Cal_UI_AD_Line.png')

plt.show()
```
