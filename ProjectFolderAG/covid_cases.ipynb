```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import gmaps
import os
from scipy.stats import linregress
from citipy import citipy
import json
from pprint import pprint 
from datetime import datetime

output_data = "output_data/clean_covid.csv"
```

```python
covid_df = pd.DataFrame(columns = ['Date'])
```

```python
states = ["ca", "tx", "sd", "md", "ga", "fl"]

index = 0
 

for state in states:
    
    url = f"https://api.covidtracking.com/v1/states/{state}/daily.json"
    
    response = requests.get(url)
    response_json = response.json()
    
    for row in response_json:
        
        try:
            print(f"Printing data for {state} and index: {index}")

            
            covid_df.loc[index, 'Date'] = row['date']
            covid_df.loc[index, 'State'] = row['state']
            covid_df.loc[index, 'Positive'] = row['positive']
            covid_df.loc[index, 'Positive Increase'] = row['positiveIncrease']
            covid_df.loc[index, 'Death Increase'] = row['deathIncrease']
            covid_df.loc[index, 'Death Confirmed'] = row['death']
            covid_df.loc[index, 'Death Probable'] = row['deathProbable']
            covid_df.loc[index, 'Hospitalized'] = row['hospitalized']
            covid_df.loc[index, 'Negative'] = row['negative']
            covid_df.loc[index, 'Recovered'] = row['recovered']
            covid_df.loc[index, 'Total'] = row['total']
            covid_df.loc[index, 'Total Test Results'] = row['totalTestResults']
           
            index += 1
           

        except (KeyError, IndexError):
            print("info not found. Skipping....")
  

```

```python
test = covid_df.copy()
```

```python
test["Date Format"] = ""
```

```python
test["Date Format"] = pd.to_datetime(test["Date"], format="%Y%m%d")
```

```python
test["weeknum"] = ""
```

```python
test["weeknum"] = test["Date Format"].dt.isocalendar().week
```

```python
test.to_csv(output_data, index=False)
```

```python
test_group_edit = test.rename(columns = {"State":"Code"})
test_group_edit
```
